```python
from roboflow import Roboflow
from dotenv import load_dotenv
import os
from ultralytics import YOLO
```

```python
load_dotenv()
```

```python
Rflow_key = os.getenv("ROBOFLOW_API_KEY")
print("Roboflow API Key loaded response:", Rflow_key is not None)
```

```python
#Accessing the project and the dataset

rf = Roboflow(api_key=Rflow_key)
project = rf.workspace("lumen-visual-assistant").project("environmental_hazards-vnsbh")

print(f"Project load Response: {project is not None}")
```

```python
#Exporting the dataset
dataset = project.version(1).download("yolov8")
print(f"Dataset creation Response: {dataset is not None}")
```

```python
#First we get dir path for ease
env_hz_dir = os.environ.get("ENV_HZ_LMN_PATH")
```

```python
#Create a new dir for result saving

save_dir = env_hz_dir + "/" + "Version1-Results"
os.makedirs(save_dir, exist_ok=True)
```

```python
#Load and train the model on the dataset (80 Epochs for starting)

model = YOLO('yolo11s.pt')

results = model.train(
    data=env_hz_dir + '/' + 'Environmental_Hazards-1' + '/' + 'data.yaml',        
    epochs=50,               
    imgsz=640,               
    project=save_dir,               
    device="0",          
    amp=True,
)
```
