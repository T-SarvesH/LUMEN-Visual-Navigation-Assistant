```python
#Testing version 1 now

from ultralytics import YOLO
import os
from dotenv import load_dotenv
from IPython.display import clear_output, Image, display
import cv2, time, subprocess
```

```python
load_dotenv()
```

```python
env_haz_path = os.environ.get("ENV_HZ_LMN_PATH")
```

```python
#Initialize the model
model_path = env_haz_path + "/" + "Version1-Results/train/weights/best.pt"
model = YOLO(model_path)
```

```python
#Testing the model on the video
cap = cv2.VideoCapture("Version1Testing-1.mp4")

frame_width  = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = 30  

# Step 1: Write temporary MKV (safe with OpenCV on Ubuntu)
temp_video = "temp_output.mkv"
out = cv2.VideoWriter(temp_video, cv2.VideoWriter_fourcc(*'MJPG'), fps, (frame_width, frame_height))

frame_count = 0
total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
print(f"Total frames to process: {total_frames}")

while True:
    ret, frame = cap.read()
    if not ret:
        break
    frame_count += 1

    results = model(frame)
    annotated_frame = results[0].plot()
    out.write(annotated_frame)

cap.release()
out.release()

# Step 2: Convert to MP4 with ffmpeg (H.264 codec for universal playback)
final_video = "Version1Testing-1(Annotated).mp4"
subprocess.run([
    "ffmpeg", "-y", "-i", temp_video, 
    "-c:v", "libx264", "-preset", "fast", "-crf", "22", 
    final_video
])

print(f"\n Annotated video saved as '{final_video}'")
```
